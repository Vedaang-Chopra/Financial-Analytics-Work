```python
Base = declarative_base()
class Mutual_Funds_Detailed_Info(Base):
    id =                    Column(name= "id", type_=Integer(), nullable=False, autoincrement=True)
    mutual_fund_name=       Column(name= "mutual_fund_name",type_=Text(), nullable=False)
    scheme_name=            Column(name= "scheme_name",type_=Text(), nullable=False, primary_key=True)
    fund_type=              Column(name= "fund_type",type_=Text(), nullable=False)
    fund_category=          Column(name= "fund_category",type_=Text(), nullable=False)
    fund_sub_category=      Column(name= "fund_sub_category",type_=Text(), nullable=False)
    isin_div_growth=        Column(name= "isin_div_growth",type_=Text(), nullable=False)
    isin_div_reinv=         Column(name= "isin_div_reinv",type_=Text(), nullable=False)
    record_date=            Column(name= "record_date",type_=DateTime(), nullable=False)
    __tablename__ = 'mutual_fund_detailed_info'
    __table_args__ = {
        'extend_existing': True,
        # "primary_key" : ['mutual_fund_nav_value.id', 'mutual_fund_nav_value.mutual_fund_id'], 
        }
Base.metadata.create_all(engine)
```
