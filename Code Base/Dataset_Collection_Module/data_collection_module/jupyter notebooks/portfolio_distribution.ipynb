```python
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
```

```python
# Set up the Selenium WebDriver
# driver_path = '/path/to/chromedriver'  # Replace with the path to your WebDriver
# driver = webdriver.Chrome(executable_path=driver_path)
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)
# URL of the AMFI portfolio disclosure page
url = 'https://www.amfiindia.com/investor-corner/online-center/portfoliodisclosure'
# Open the page using Selenium
driver.get(url)
```

```python
# Wait for the dropdown menu to appear
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "ui-id-1"))
)
```

```python
# Wait for the dropdown to load and click on the "Monthly Portfolio Disclosure" option
WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.ID, "ui-id-1"))  # ID for "Monthly Portfolio Disclosure"
).click()
```

```python
# Make the dropdown menu visible if necessary (simulate user action)
dropdown_trigger = driver.find_element(By.ID, "ui-id-1")  # Replace with actual ID or class
dropdown_trigger.click()
```

```python
# Locate the "Monthly Portfolio Disclosure" option
monthly_option = driver.find_element(By.XPATH, '//*[@id="ui-id-1"]/li[2]')

# Click the option to generate the table
monthly_option.click()
```

```python
# Wait for the table to load (adjust the timeout as needed)
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.TAG_NAME, "table"))
)
```

```python
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find the table by its class (replace 'table-class-name' with the actual class of the table)
table = soup.find('table', {'class': 'table-class-name'})  # Replace with the actual table class

# Extract all <td> links based on a particular class (replace 'td-class-name' with the actual class)
td_links = table.find_all('td', {'class': 'td-class-name'})  # Replace with the actual td class
```

```python
# Initialize a list to store the extracted links
links = []

# Iterate over the <td> elements to extract the links
for td in td_links:
    link = td.find('a', href=True)  # Find the <a> tag within the <td>
    if link:
        links.append(link['href'])

# Print the extracted links
for link in links:
    print(link)
```

```python
# Iterate over each option to simulate a click
for option in options:
    # Click on the option
    option.click()
    time.sleep(3)  # Wait for the content to load

    # Scrape the dynamically loaded content
    # soup = BeautifulSoup(driver.page_source, 'html.parser')
    print(option)
```

```python
# Get the page source and parse it with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python







# Find the section containing the monthly portfolio disclosures
monthly_disclosure_section = soup.find('div', {'id': 'divMonthlyPortfolio'})

if monthly_disclosure_section:
    # Extract all anchor tags within the section
    anchor_tags = monthly_disclosure_section.find_all('a', href=True)

    # Initialize a list to store AMC names and URLs
    amc_list = []

    # Iterate over each anchor tag to extract AMC name and URL
    for anchor in anchor_tags:
        amc_name = anchor.get('title', 'No Title').strip()
        amc_url = anchor['href'].strip()
        if amc_name and amc_url:
            amc_list.append({'AMC Name': amc_name, 'Website': amc_url})

    # Display the extracted AMC names and URLs
    for amc in amc_list:
        print(f"AMC Name: {amc['AMC Name']}, Website: {amc['Website']}")
else:
    print("Could not find the monthly portfolio disclosure section.")

# Close the browser
driver.quit()
```
